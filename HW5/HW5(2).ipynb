```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_df = pd.read_csv('ToyotaCorolla.csv')
data_df.head()
```

```python
predictors = ['Age_08_04','KM','HP','Automatic','Doors','Quarterly_Tax','Mfr_Guarantee','Guarantee_Period','Airco','Automatic_airco','CD_Player','Powered_Windows','Sport_Model','Tow_Bar','Fuel_Type']
X = data_df[predictors]
X = pd.get_dummies(X, columns=['Fuel_Type'])
y = data_df['Price']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.4, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

regressor = DecisionTreeRegressor()

regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_valid)
mse = mean_squared_error(y_valid, y_pred)
print("Mean Squared Error:", mse)
```
