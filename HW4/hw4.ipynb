```python
import pandas as pd
```

```python
accidents_df = pd.read_csv('accidentsFull.csv')
accidents_df.head()
```

```python
accidents_df['Injury'] = accidents_df['MAX_SEV_IR'].apply(lambda x: 1 if x in [1, 2] else 0)
```

```python
df_head = accidents_df.head(12)
df_head.shape
predictors = ['WEATHER_R','TRAF_CON_R']
```

```python
for predictor in predictors:
    df = df_head[['Injury', predictor]]
    pivot_table = df.pivot_table(df, index='Injury', columns=predictor, aggfunc=len, fill_value=0)
    propTable = pivot_table.apply(lambda x: x /sum(x), axis=1)
    print(propTable)
```
